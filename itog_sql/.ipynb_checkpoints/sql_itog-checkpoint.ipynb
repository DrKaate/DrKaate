## Задачи

### Cколько книг вышло после 1 января 2000 года

```python
query = '''

SELECT COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01'

'''

pd.io.sql.read_sql(query, con = engine)
```

После 1 января 2000 года вышло 819 книг и брошюр

### Для каждой книги посчитаем количество обзоров и среднюю оценку

```python
query = '''

SELECT b.book_id, 
       b.title, 
       COUNT(DISTINCT rw.review_id) AS cnt_reviews, 
       ROUND(AVG(rg.rating),3) AS avg_rating
FROM books AS b

LEFT JOIN reviews AS rw
ON b.book_id = rw.book_id

LEFT JOIN  ratings AS rg
ON b.book_id = rg.book_id

GROUP BY b.book_id, b.title
ORDER BY cnt_reviews DESC

'''

pd.io.sql.read_sql(query, con = engine)
```

Мы получили данные по 1000 книг, посчитали количество обзоров для каждой и среднюю оценку по результатам этих обзоров.

### Определим издательство, которое выпустило наибольшее число книг толще 50 страниц 

```python
query = ''' 

SELECT publisher, 
       COUNT(book_id) as books_count 
FROM books AS b
LEFT JOIN publishers p ON b.publisher_id = p.publisher_id
WHERE num_pages > 50
GROUP BY publisher
ORDER BY books_count DESC
LIMIT 1            

'''
pd.io.sql.read_sql(query, con = engine)
```

Применив фильтрацию по количеству страниц мы исключили из анализа брошюры, остались толькол книги. Наибольшее количество книг (в количестве `42`) выпустило издательство `Penguin Books`.	

### Определим автора с самой высокой средней оценкой книг — учитывайть будем только книги с 50 и более оценками

```python
query = '''

WITH 
fifty AS (
    SELECT book_id
    FROM ratings
    GROUP BY book_id
    HAVING COUNT(rating_id) >= 50)


SELECT a.author_id, 
       a.author, 
       AVG(r.rating) AS avg_rating
FROM books AS b

INNER JOIN fifty f
ON b.book_id = f.book_id

INNER JOIN authors a
ON b.author_id = a.author_id

INNER JOIN ratings r
ON f.book_id = r.book_id

GROUP BY a.author_id, a.author
ORDER BY AVG(r.rating) DESC

LIMIT 1

'''

pd.io.sql.read_sql(query, con = engine)
```

Сначала мы создали временную таблицу, где отобрали книги с количеством оценок не менее 50.
Затем определили автора с самой высокой оценкой книг – `J.K. Rowling/Mary GrandPré` со средней оценкой `4.29`

### Выполним рассчет среднего количества обзоров от пользователей, которые поставили больше 50 оценок

```python
query = '''

SELECT COUNT(DISTINCT rw.text)/ COUNT(DISTINCT rw.username) AS avg_rv_count
FROM ratings AS r

INNER JOIN reviews AS rw
ON r.username = rw.username

WHERE r.username in ( 
                    SELECT username
                    FROM ratings
                    GROUP BY username
                    HAVING COUNT(rating_id) > 50
                    )

'''

pd.io.sql.read_sql(query, con = engine)
```

Пользователи, поставившие более 50 оценок, в среднем, выполнили по 24 обзора
